```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import the insurance data & check to see what it looks like
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
# plotting a scatter plot 
x = insurance_data.loc[:, 'age'].values
y = insurance_data.loc[:, 'charges'].values

plt.scatter(x, y)
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
# fit the model for linear regression
insurance_model = LinearRegression()
insurance_model.fit(insurance_data[['age']].values, insurance_data['charges'].values)
```

```python
# make a prediction & plot another scatter plot with the best-fit line
predict = insurance_model.predict(insurance_data[['age']].values)

plt.scatter(insurance_data['age'], predict)
plt.plot(insurance_data['age'], predict, color = 'r')
plt.xlabel('age')
plt.ylabel('Predicted Charges')
plt.show
```

```python
# makeing a prediction on insurance charges if the age was 85
age_int = 85
new_age = [[age_int]]

age_predict = np.append(insurance_data['age'], new_age).reshape(-1,1)
charges_predict = insurance_model.predict(age_predict)

plt.scatter(insurance_data['age'],insurance_data['charges'])
plt.plot(age_predict, charges_predict, color = 'r')
plt.xlabel('age')
plt.ylabel('Predicted Charges')
plt.show

formatted_charges = "£{:,.2f}".format(float(insurance_model.predict(new_age)))
print(f"Estimated predicted insurance charges for an {age_int} year old: {formatted_charges}")
```
