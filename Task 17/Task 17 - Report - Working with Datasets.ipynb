```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt',sep=' ')
```

```python
# compare average income based on ethnicity
df.groupby('Ethnicity')['Income'].mean()
```

```python
# compare married or single people average balance
married_average = df[df.Married == "Yes"].loc[:,"Balance"].mean()
single_average = df[df.Married == "No"].loc[:,"Balance"].mean()

print("Married People Average =", married_average)
print("Single People Average =", single_average)

# Compare average balances and print the result
if single_average > married_average:
    print("Single People have a higher average balance")
elif single_average < married_average:
    print("Married People have a higher average balance")
else:
    print("Average balances for single and married people are equal")
```

```python
# Highest income in Dataset
print("Highest income in dataset =", df.loc[:, "Income"].max())
```

```python
# Lowest income in Dataset
print("Lowest income in dataset =", df.loc[:, "Income"].min())
```

```python
# Total cards recorded in dataset
print("Total cards in dataset =", df.loc[:, "Cards"].sum())
```

```python
# Count of Female & Male info in dataset
df.groupby('Gender')['Gender'].count()
```
